```python
def place_order(customer_name, restaurant, item="Pizza", *extras, delivery_time="ASAP", **delivery_details):
    print(f"Order placed by: {customer_name}")
    print(f"Restaurant: {restaurant}")
    print(f"Main item: {item}")

    # Handling extras (*args)
    if extras:
        print("Extras:", ", ".join(extras))

    # Handling default argument
    print(f"Delivery Time: {delivery_time}")

    # Handling **kwargs
    if delivery_details:
        print("Delivery Details:")
        for key, value in delivery_details.items():
            print(f"  {key}: {value}")

    print("-" * 40)
# 1. Positional arguments
place_order("Alice", "Domino's")

# 2. Default argument (item defaults to Pizza)
place_order("Bob", "KFC")

# 3. Keyword arguments (changing delivery_time)
place_order("Charlie", "Burger King", item="Burger", delivery_time="7 PM")

# 4. *args (multiple extra items)
place_order("Diana", "Subway", "Sandwich", "Coke", "Chips")

# 5. **kwargs (extra delivery info)
place_order("Eve", "Pizza Hut", "Pasta", "Garlic Bread", delivery_time="8 PM",
            address="123 Main St", phone="9876543210", payment="Online")
```

```python
import matplotlib.pyplot as plt

population = [100]
growth_rate = 1.05  # 5% growth per year

for year in range(1, 21):
    population.append(population[-1] * growth_rate)

plt.plot(population)
plt.xlabel("Years")
plt.ylabel("Population")
plt.title("Population Growth Prototype")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Parameters
population = 1000
infected = 1
recovered = 0
beta = 0.3   # infection rate
gamma = 0.1  # recovery rate
days = 100

S, I, R = [population - infected], [infected], [recovered]

# Simulation
for _ in range(days):
    new_infected = beta * S[-1] * I[-1] / population
    new_recovered = gamma * I[-1]

    S.append(S[-1] - new_infected)
    I.append(I[-1] + new_infected - new_recovered)
    R.append(R[-1] + new_recovered)

# Plot results
plt.plot(S, label="Susceptible")
plt.plot(I, label="Infected")
plt.plot(R, label="Recovered")
plt.xlabel("Days")
plt.ylabel("Number of people")
plt.title("Disease Spread Simulation (Prototype)")
plt.legend()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Load dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize & preprocess
x_train, x_test = x_train / 255.0, x_test / 255.0
y_train, y_test = to_categorical(y_train), to_categorical(y_test)

# Quick prototype model
model = Sequential([
    Flatten(input_shape=(28, 28)),   # 28x28 image  flat vector
    Dense(128, activation='relu'),   # hidden layer
    Dense(10, activation='softmax')  # 10 classes (digits 09)
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=3, batch_size=64, validation_split=0.1)

# Evaluate
loss, acc = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {acc:.2f}")
```

```python
def parcel_scanning_system(parcel_code: list[str]) -> list[str]:
    messages = []
    for code in parcel_code:
        if code == "DAMAGED":
            messages.append("Skipped damaged parcel")
            continue
        if code == "STOP":
            messages.append("Critical error: Stopping scan")
            break
        messages.append(f"Scanned parcel: {code}")
    else:
        messages.append("All parcels scanned successfully")
    return messages
parcel_code = ["12345", "DAMAGED", "67890", "STOP", "11111"]
print(parcel_scanning_system(parcel_code))
```

```python
import json

# Sample JSON data (this would normally come from a file)
data_json = '''
{
  "orders": [
    {"id": 101, "customer": "Alice", "items": 3, "amount": 250.0, "status": "pending"},
    {"id": 102, "customer": "Bob", "items": 1, "amount": 120.5, "status": "pending"},
    {"id": 103, "customer": "Charlie", "items": 2, "amount": 450.0, "status": "approved"}
  ]
}
'''

# Parse JSON
orders_data = json.loads(data_json)["orders"]

# Function 1: Generate Report
def generate_report(orders):
    print("=== Report Generated ===")
    print(f"Total Orders: {len(orders)}")

# Function 2: Fetch Sales
def fetch_sales(orders):
    total_sales = sum(order["amount"] for order in orders)
    print(f"Total Sales Value: ${total_sales:.2f}")

# Function 3: Validate Order
def validate_order(orders):
    for order in orders:
        if order["status"] == "pending":
            print(f"Order {order['id']} is pending validation.")
        else:
            print(f"Order {order['id']} already validated.")

# Function 4: Summarize Order
def summarize_order(orders):
    for order in orders:
        print(f"Order {order['id']} | Customer: {order['customer']} | Items: {order['items']} | Amount: ${order['amount']:.2f}")

# Function 5: Close Report
def close_report():
    print("=== Report Closed ===")

# Calling functions with real use case
generate_report(orders_data)
fetch_sales(orders_data)
validate_order(orders_data)
summarize_order(orders_data)
close_report()
```

```python
# Multiply Length

def list_multi(my_list):
  result = 1
  for num in my_list:
    result *= num
  return result

my_list = [10,20,30,40,50]
result = list_multi(my_list)
print(f"Multiplication of list is {result}")
```

```python
# Find largest n number in list

def largest_number(my_list):
  largest = my_list[0]
  for num in my_list:
    if num > largest:
      largest = num
  return largest

my_list = [10,20,30,40,50]
result = largest_number(my_list)
print(f"Largest number is {result}")
```

```python
# List Negative number

def find_negative(my_list):
  negative_list = []
  for num in my_list:
    if num < 0:
      negative_list.append(num)
  return negative_list

my_list = [10,-20,30,-40,50]
result = find_negative(my_list)
print(f"Negative number is {result}")
```

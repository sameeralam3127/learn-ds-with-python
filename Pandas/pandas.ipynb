##  What Is Pandas?

**Pandas** (Python Data Analysis Library) is a **powerful, fast, flexible** open-source tool for:

* **Data manipulation**
* **Data cleaning**
* **Data analysis**
* **Data visualization (basic)**

It provides two core data structures:

1. **Series** – 1D labeled array (like a column in Excel)
2. **DataFrame** – 2D labeled table (like an Excel sheet or SQL table)
